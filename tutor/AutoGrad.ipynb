# Автоматичне диференціювання з AutoGrad

Замість обчислення градієнта «вручну», ми використовуємо набір інструментів для автоматичного диференціювання, в даному випадку [AutoGrad](https://github.com/HIPS/autograd).

AutoGrad використовує автоматичне диференціювання в зворотному режимі для обчислення градієнта з тими ж витратами, що і для обчислення самої функції. Дивіться [цю статтю](https://rufflewind.com/2016-12-30/reverse-mode-automatic-differentiation) для отримання додаткової інформації.

```python
import autograd.numpy as np
import autograd as ag
from autograd import elementwise_grad as egrad

import matplotlib.pyplot as plt
```

Тест на скалярній функції, обчислення та побудова графіків похідних вищого порядку.

```python
def tanh(x):
    return (1.0 - np.exp(-x))  / (1.0 + np.exp(-x))
t = np.linspace(-7, 7, 200)
plt.plot(t, tanh(t),
         t, egrad(tanh)(t),                                     # first derivative
         t, egrad(egrad(tanh))(t),                              # second derivative
         t, egrad(egrad(egrad(tanh)))(t),                       # third derivative
         t, egrad(egrad(egrad(egrad(tanh))))(t),                # fourth derivative
         t, egrad(egrad(egrad(egrad(egrad(tanh)))))(t),         # fifth derivative
         t, egrad(egrad(egrad(egrad(egrad(egrad(tanh))))))(t));  # sixth derivative
```
